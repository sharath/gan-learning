```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from utils import load_data
from models import Generator, Discriminator
from samplers import UniformDatasetSampler, UniformLatentSampler, NormalLatentSampler
from config import params_wgangp
```

```python
configuration = 'wgangp'
params = eval(f'params_{configuration}')
```

```python
np.random.seed(params['seed'])
torch.manual_seed(params['seed'])
torch.cuda.manual_seed_all(params['seed'])
```

```python
data, labels = load_data('grid')
```

```python
data_sampler = UniformDatasetSampler(torch.tensor(data).float())
noise_sampler = UniformLatentSampler(params['latent_dim'])
```

```python
G = Generator(params['latent_dim'], params['model_dim'], params['data_dim'], spec_norm=params['spec_norm_g']).to(params['device'])
D = Discriminator(params['model_dim'], params['data_dim'], spec_norm=params['spec_norm_d']).to(params['device'])
```

```python
D_optimizer = optim.Adam(D.parameters(), lr=params['learning_rate'], betas=params['betas'])
G_optimizer = optim.Adam(G.parameters(), lr=params['learning_rate'], betas=params['betas'])
```

```python
def compute_gp(real_data, fake_data):
    alpha = torch.rand(real_data.shape[0], 1).expand(real_data.size()).to(params['device'])
    interpolate = (alpha * real_data + (1 - alpha) * fake_data).requires_grad_(True)
    interpolate_pred = D(interpolate)

    gradients = torch.autograd.grad(outputs=interpolate_pred,
                                    inputs=interpolate,
                                    grad_outputs=torch.ones(interpolate_pred.size()).to(params['device']),
                                    create_graph=True, retain_graph=True, only_inputs=True)[0]

    gradient_penalty = (gradients.norm(2, dim=1) - 1).pow(2).mean() * params['gp_lambda']
    return gradient_penalty
```

```python
losses = {
    'D':[],
    'G':[]
}
```

```python
D.train()
G.train()
for it in range(params['iterations']):
    D.zero_grad()

    # compute real data loss
    x_real = data_sampler.get_batch(params['batch_size']).to(params['device'])
    real_pred = D(x_real)

    # compute fake data loss
    z_fake = noise_sampler.get_batch(params['batch_size']).to(params['device'])
    x_fake = G(z_fake).detach()
    fake_pred = D(x_fake)
    
    # backpropagate and step
    gradient_penalty = compute_gp(x_real, x_fake)
    D_loss = fake_pred.mean() - real_pred.mean() + gradient_penalty
    D_loss.backward()
    D_optimizer.step()
    
    # for bookkeeping
    D_loss = fake_pred.mean() - real_pred.mean()
    
    if it % params['n_critic'] == 0:
        G.zero_grad()
        
        # compute loss

        z_fake = noise_sampler.get_batch(params['batch_size']).to(params['device'])
        x_fake = G(z_fake)
        fake_pred = D(x_fake)

        # backpropagate and step
        G_loss = -fake_pred.mean()
        G_loss.backward()
        G_optimizer.step()
                
    losses['D'].append(D_loss.item())
    losses['G'].append(-G_loss.item())
```

```python
plt.plot(losses['G'], label='Generator')
plt.plot(losses['D'], label='Discriminator')
plt.legend()
plt.show()
```
